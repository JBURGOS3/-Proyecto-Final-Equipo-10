¿Cómo se distribuyen los datos en función de diferentes categorías?

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Seleccionamos las columnas relevantes para el clustering
columns_for_clustering = ['ValueA1', 'ValueA2', 'ValueP1', 'ValueP2', 'Flow']

# Los datos relevantes para el clustering
data_for_clustering = df[columns_for_clustering]

# Aplica K-means
kmeans = KMeans(n_clusters=3)  # número de clusters
df['Cluster'] = kmeans.fit_predict(data_for_clustering)

# Visualiza los clusters en un gráfico de dispersión 2D
plt.scatter(df['ValueA1'], df['ValueA2'], c=df['Cluster'], cmap='viridis')
plt.title('Visualización de Clusters (K-means)')
plt.xlabel('ValueA1')
plt.ylabel('ValueA2')
plt.show()
```

Si . Se puede observar que los datos se pueden agrupar basandose en  los acelerometros.La visualización con el gráfico de dispersión 2D  permite observar cómo las instancias se agrupan en función de las columnas 'ValueA1' y 'ValueA2', dandonos una idea de la estructura de los datos y cómo las mediciones se relacionan entre sí en términos de aceleración


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualización de la distribución de clases con un countplot
sns.countplot(x='Leak', data=df)
plt.title('Distribución de Clases en la Variable "Leak"')
plt.xlabel('Clases')
plt.ylabel('Número de Instancias')
plt.show()

# Métricas de desbalance
class_counts = df['Leak'].value_counts()
class_proportions = class_counts / len(df)
class_imbalance_factor = class_proportions.max() / class_proportions.min()

print("Proporción de Clases:")
print(class_proportions)
print("\nFactor de Desequilibrio:")
print(class_imbalance_factor)
```

¿Hay desequilibrio en las clases de la variable objetivo? Todas las clases ("CC", "LC", y "NL") tienen una proporción de 0.333333 (aproximadamente un tercio). Esto indica que cada clase representa aproximadamente el mismo porcentaje del conjunto de datos.

El factor de desequilibrio es 1.0, Indica no hay un desequilibrio significativo entre las clases. Todas las clases tienen la misma proporción, y ninguna es significativamente más frecuente que las demás. Las clases en están distribuidas de manera equitativa.
